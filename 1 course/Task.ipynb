```python
# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import gamma
from scipy.stats import norm
%matplotlib inline
```

```python
# Генерируем выборку из 1000 элементов, рисуем гистограмму и теоретическую плотность распределения.
a = 1.89
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')
x = np.linspace(0,10, 1000)
rv = gamma(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='theoretical pdf')
r = gamma.rvs(a, size=1000)
ax.hist(r, bins=30, normed=True, histtype='stepfilled', alpha=0.5)

ax.legend(loc='best', frameon=False)
```

```python
# Формируем выборки объема n с помощью size
def samples(n):
    s_mean = []
    for i in range(0,1000):
        r = gamma.rvs(a, size = n)
        s_mean = np.append(s_mean, r.mean())
    return s_mean 
```

```python
# Строим гистограмму выборочных средний и график нормального распределения для n=5
n = 5
s_mean = samples(n)
plt.hist(s_mean, bins=30, normed=True, histtype='stepfilled', alpha=0.5, label='s_mean')
norm_x = norm(s_mean.mean(), np.sqrt(var/n))
plt.plot(x, norm_x.pdf(x), 'k-', lw=2, label='norm pdf')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
