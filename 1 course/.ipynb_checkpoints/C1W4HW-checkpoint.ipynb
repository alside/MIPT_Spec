```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import gamma
%matplotlib inline
```

# Центральная предельная теорема своими руками

Для доказательства теоремы было выбрано gamma-распределение

```python
#инициализируем генератор, генерируем выборку из 1000 элементов
#отображаем выборку в виде гистограммы 
# и накладываем на нее теоретическую плотность распределения
a = 1.89
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')
x = np.linspace(0,10, 1000)
rv = gamma(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='theoretical pdf')
r = gamma.rvs(a, size=1000)
ax.hist(r, bins=30, normed=True, histtype='stepfilled', alpha=0.5)

ax.legend(loc='best', frameon=False)
```

```python
# формирование 1000 выборок объема n и подсчет их средних
# возвращает массив средних
from scipy.stats import norm
def nsamples(n):
    s_mean = []
    for i in range(0,1000):
        r = gamma.rvs(a, size = n)
        s_mean = np.append(s_mean, r.mean())
    return s_mean 
```

```python
# строим гистограмму выборочных средних 
# накладываем на гистограмму график нормального распределения
# параметры нормального распределения считаются как mean исходной выборки и sqrt(variance/n)
n = 5
s_mean = nsamples(n)
plt.hist(s_mean, bins=30, normed=True, histtype='stepfilled', alpha=0.5, label='s_mean')
norm_x = norm(s_mean.mean(), np.sqrt(var/n))
plt.plot(x, norm_x.pdf(x), 'k-', lw=2, label='norm pdf')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# строим гистограмму выборочных средних 
# накладываем на гистограмму график нормального распределения
# параметры нормального распределения считаются как mean исходной выборки и sqrt(variance/n)
n = 10
s_mean = nsamples(n)
plt.hist(s_mean, bins=30, normed=True, histtype='stepfilled', alpha=0.5, label='s_mean')
norm_x = norm(s_mean.mean(), np.sqrt(var/n))
plt.plot(x, norm_x.pdf(x), 'k-', lw=2, label='norm pdf')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# строим гистограмму выборочных средних 
# накладываем на гистограмму график нормального распределения
# параметры нормального распределения считаются как mean исходной выборки и sqrt(variance/n)
n=100
s_mean = nsamples(n)
plt.hist(s_mean, bins=30, normed=True, histtype='stepfilled', alpha=0.5, label='s_mean')
norm_x = norm(s_mean.mean(), np.sqrt(var/n))
plt.plot(x, norm_x.pdf(x), 'k-', lw=2, label='norm pdf')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

При увеличении числа n, расперделения все больше становятся похожи на нормальное, т.е. точность апроксимации возрастает с ростом параметра n.
